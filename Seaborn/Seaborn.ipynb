```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Barplot

```python
# Creating a barplot between cyl and mpg of mtcars dataset
sns.barplot(x = 'cyl', y = 'mpg', data = mtcars)
plt.show()
```

```python
# Changing the color of bars using color parameter
sns.barplot(x = 'cyl', y = 'mpg', data = mtcars, color = 'orange')
plt.show()
```

```python
# Using Color Palette 
sns.barplot(x = 'am', y = 'mpg', data = mtcars, palette = 'Set2')
plt.show()
```

```python
sns.barplot(x = 'gear', y = 'hp', data = mtcars, palette = 'rocket')
plt.show()
```

## Countplot

```python
sns.countplot(x = 'cyl', data = mtcars, palette = "Set1")
plt.show()
```

```python
sns.countplot(x = "carb", data = mtcars, palette = "rocket")
plt.show()
```

### Horizontal Countplot

```python
sns.countplot(y = 'gear', data = mtcars, palette = "rocket")
plt.show()
```

### Grouped Count Plot (Hue Parameter)

```python
sns.countplot(x = 'gear', hue = 'cyl', data = mtcars)
plt.show()
```

## Displot

```python
sns.displot(mtcars.mpg, bins = 10, color = "r")
plt.show()
```

## Heat Maps

```python
sns.heatmap(mtcars.corr(), cbar = True, linewidth = 0.5)
plt.show()
```

## Scatterplot

```python
iris = pd.read_csv("C:/Users/yashr/Desktop/Movies/iris.csv")
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
sns.scatterplot(x = 'Sepal.Length', y = 'Petal.Length', data = iris)
plt.show()
```

```python
sns.scatterplot(x = 'Sepal.Length', y = 'Petal.Length', data = iris, hue = "Species", size = "Species")
plt.show()
```

## Pairplot

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue = "Species")
plt.show()
```

## Linear Model Plot

```python
sns.lmplot(x ="Petal.Length", y ="Petal.Width", data=iris)
plt.show()
```

```python
sns.lmplot(x = "Petal.Length", y = "Petal.Width", hue = "Species", data = iris, markers = ["o", "*", "^"])
plt.show()
```

## Boxplot

```python
sns.boxplot(x = "Species", y = "Sepal.Width", data = iris)
plt.show()
```
