## Pandas - Cleaning Empty Cells
### Empty Cells
Empty cells can potentially give you a wrong result when you analyze data.<br>
One way to deal with empty cells is to remove rows that contain empty cells.

```python
# Dataset
# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv("data1.csv")

# Printing new dataframe
df
```

```python
# Return a new Data Frame with no empty cells
# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv("data1.csv")

# Dropping the empty cells
new_df = df.dropna()

# Printing new dataframe
new_df
```

By default, the <span style = "color: red"><b>dropna()</b></span> method returns a new DataFrame, and will not change the original.<br>
If we want to change the original DataFrame, use the <span style = "color: red"><b>inplace = True</b></span> argument

```python
# Return a new Data Frame with no empty cells
# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv("data1.csv")

# Dropping the empty cells
df.dropna(inplace = True)

# Printing dataframe
df
```

The <span style = "color: red"><b>dropna(inplace = True)</b></span> will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

### Replace Empty Values
1. Another way of dealing with empty cells is to insert a new value instead.<br>
1. This way you do not have to delete entire rows just because of some empty cells.<br>
1. The <span style = "color: red"><b>fillna()</b></span> method allows us to replace empty cells with a value

```python
# Replace NULL values with the number 130

# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv('data1.csv')

# Replacing the empty cells with value 130
df.fillna(130, inplace = True)  

# Printing the dataframe
df
```

We can see that row 22 date value 'na' is also replaced by 130. so we need to be careful while replacing the value.

### Replace Only For Specified Columns
To only replace empty values for one column, specify the column name for the DataFrame

```python
# Replace NULL values in the "Calories" columns with the number 130

# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv('data1.csv')

# Replacing NULL vaues of calories with number 130
df['Calories'].fillna(130, inplace = True)

# Printing the dataframe
df
```

### Replace Using Mean, Median, or Mode
A common way to replace empty cells, is to calculate the mean, median or mode value of the column.<br>
Pandas uses the <span style = "color: red"><b>mean()</b></span> <span style = "color: red"><b>median()</b></span> and <span style = "color: red"><b>mode()</b></span> methods to calculate the respective values for a specified column


```python
# Calculate the MEAN, and replace any empty values with it

# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv('data1.csv')

# Calculating mean of calories
mean_cal = df['Calories'].mean()

# Replacing the NULL values with mean
df['Calories'].fillna(mean_cal, inplace = True)

# Printing the dataframe
df
```

<span style = "color: red; background-color : yellow"><b>Mean</b></span> = the average value (the sum of all values divided by number of values).

```python
# Calculate the MEDIAN, and replace any empty values with it

# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv('data1.csv')

# Calculating median of calories
median_cal = df['Calories'].median()

# Replacing the NULL values with median
df['Calories'].fillna(median_cal, inplace = True)

# Printing the dataframe
df
```

<span style = "color: red; background-color : yellow"><b>Median</b></span> = the value in the middle, after you have sorted all values ascending.

```python
# Calculate the MODE, and replace any empty values with it:

# Importing pandas
import pandas as pd

# Loading dataset into dataframe
df = pd.read_csv('data1.csv')

# Calculating mean of calories
mode_cal = df['Calories'].mode()[0]

# Replacing the NULL values with mode
df['Calories'].fillna(mode_cal, inplace = True)

# Printing the dataframe
df
```

<span style = "color: red; background-color : yellow"><b>Mode</b></span> = the value that appears most frequently.

## Pandas - Fixing Wrong Data
### Wrong Data
1. "Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".<br>
1. Sometimes you can spot wrong data by looking at the data set, because you have an expectation of what it should be.<br>
1. If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.<br>
1. It doesn't have to be wrong, but taking in consideration that this is the data set of someone's workout sessions, we conclude with the fact that this person did not work out in 450 minutes.

### Replacing Values
In our example, it is most likely a typo, and the value should be "45" instead of "450", and we could just insert "45" in row 7

```python
# Set "Duration" = 45 in row 7

# Importing pandas
import pandas as pd

# Loading The dataset Date.csv
df = pd.read_csv('Date.csv')

# Converting Date Column into Date using to_datetime() method
df['Date'] = pd.to_datetime(df['Date'], format="mixed")

# Dropping the NaT Value
df.dropna(subset = ['Date'], inplace = True)

# Dropping the NaN value row from Calories
df.dropna(subset = ['Calories'], inplace = True)

# Setting Duration of wrong data 45 in row 7
df.loc[7, 'Duration'] = 45

# Printing the dataframe
df
```

We can see above that row 7 has been changed

To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.

```python
# Importing pandas
import pandas as pd

# Loading The dataset Date.csv
df = pd.read_csv('Date.csv')

# # Converting Date Column into Date using to_datetime() method
df['Date'] = pd.to_datetime(df['Date'], format='mixed')

# # Dropping the NaT Value
df.dropna(subset = ['Date'], inplace = True)

# # Dropping the NaN value row from Calories
df.dropna(subset = ['Calories'], inplace = True)

# Loop through all values in the "Duration" column.
# If the value is higher than 120, set it to 120
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.loc[x, 'Duration'] = 120
# Printing the dataframe
df
```

We can see above that row 7 Duration value is changed to 120

### Removing Rows
Another way of handling wrong data is to remove the rows that contains wrong data.<br>
This way we do not have to find out what to replace them with, and there is a good chance yweou do not need them to do our analyses.

```python
# Importing pandas
import pandas as pd

# Loading The dataset Date.csv
df = pd.read_csv('Date.csv')

# Converting Date Column into Date using to_datetime() method
df['Date'] = pd.to_datetime(df['Date'])

# Dropping the NaT Value
df.dropna(subset = ['Date'], inplace = True)

# Dropping the NaN value row from Calories
df.dropna(subset = ['Calories'], inplace = True)

# Loop through all values in the "Duration" column.
# If the value is higher than 120, then remove the row
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.drop(x, inplace = True)

# Printing the dataframe
df
```

We can see above that row 7 have been removed

## Pandas - Removing Duplicates
To discover duplicates, we can use the <span style = "color: red; background-color : yellow"><b>duplicated()</b></span> method.<br>
The <span style = "color: red; background-color : yellow"><b>duplicated()</b></span> method returns a Boolean values for each row

```python
# Importing pandas
import pandas as pd

# Loading The dataset Date.csv
df = pd.read_csv('Date.csv')

# Returns True for every row that is a duplicate, othwerwise False
df.duplicated()
```

We can see that row 12 is duplicate

### Removing Duplicates
To remove duplicates, use the <span style = "color: red; background-color : yellow"><b>drop_duplicates()</b></span> method.

```python
# Importing pandas
import pandas as pd

# Loading The dataset Date.csv
df = pd.read_csv('Date.csv')

# Converting Date Column into Date using to_datetime() method
df['Date'] = pd.to_datetime(df['Date'])

# Dropping the NaT Value
df.dropna(subset = ['Date'], inplace = True)

# Dropping the NaN value row from Calories
df.dropna(subset = ['Calories'], inplace = True)

# Loop through all values in the "Duration" column.
# If the value is higher than 120, then remove the row
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.drop(x, inplace = True)

# Removing all duplicates
df.drop_duplicates(inplace = True)

# Printing the dataframe
df
```

We can see that row 12 have been removed

## Pandas - Data Correlations
### Finding Relationships
A great aspect of the Pandas module is the <span style = "color: red; background-color : yellow"><b>corr()</b></span> method.<br>
The <span style = "color: red; background-color : yellow"><b>corr()</b></span> method calculates the relationship between each column in your data set.<br>
Note: The <span style = "color: red; background-color : yellow"><b>corr()</b></span> method ignores "not numeric" columns.

```python
# Show the relationship between the columns

# Importing pandas 
import pandas as pd

# Loading the dataset into dataframe
df = pd.read_csv('Date.csv')

# Showing the relationship
print(df.corr())
```

### Result Explained
The Result of the <span style = "color: red; background-color : yellow"><b>corr()</b></span> method is a table with a lot of numbers that represents how well the relationship is between two columns.<br>
1. The number varies from -1 to 1.<br>
2. 1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.<br>
3. 0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.<br>
4. -0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.<br>
5. 0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

#### What is a good correlation? 
It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.

#### Perfect Correlation:
We can see that "Duration" and "Duration" got the number 1.000000, which makes sense, each column always has a perfect relationship with itself.

#### Good Correlation:
"Duration" and "Calories" got a 0.922721 correlation, which is a very good correlation, and we can predict that the longer you work out, the more calories you burn, and the other way around: if you burned a lot of calories, you probably had a long work out.

#### Bad Correlation:
"Duration" and "Maxpulse" got a 0.009403 correlation, which is a very bad correlation, meaning that we can not predict the max pulse by just looking at the duration of the work out, and vice versa.

The histogram tells us that there were over 100 workouts that lasted between 50 and 60 minutes.
